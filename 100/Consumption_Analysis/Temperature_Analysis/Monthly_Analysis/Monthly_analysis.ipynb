```python
import pandas as pd

# ==========================================
# LOAD DATASET
# ==========================================

# Replace with your actual file path
df = pd.read_excel(r"C:\Users\Amey\OneDrive - College of Engineering Pune(An autonomous Institute of Govt. of Maharashtra)\Desktop\Amey\Python\100\100_Pre_done_Combined.xlsx")

# ==========================================
# BASIC CLEANING
# ==========================================

df['Year'] = df['Year'].astype(int)
df['Month'] = df['Month'].astype(int)

# Convert Quantity to absolute value
df['Failure_Count'] = df['Quantity'].abs()

# ==========================================
# ==========================================================
# 1. MONTHLY ALL SLocs (WHOLE INDIA)
# ==========================================================
# Covers:
# - All SLocs
# - Whole India
# - Monthly from 2020 to 2026
# ==========================================================

monthly_india = (
    df.groupby(['Year', 'Month'])
    .agg(
        Avg_Temp=('Tavg', 'mean'),
        Failure_Count=('Failure_Count', 'sum')
    )
    .reset_index()
)

print("\nMONTHLY ALL SLocs - WHOLE INDIA")
print(monthly_india)

# ==========================================
# ==========================================================
# 2. MONTHLY REGIONWISE
# ==========================================================
# Covers:
# - All SLocs inside each region
# - Monthly from 2020 to 2026
# ==========================================================

monthly_region = (
    df.groupby(['Region', 'Year', 'Month'])
    .agg(
        Avg_Temp=('Tavg', 'mean'),
        Failure_Count=('Failure_Count', 'sum')
    )
    .reset_index()
)

print("\nMONTHLY REGIONWISE")
print(monthly_region)

# ==========================================
# ==========================================================
# 3. MONTHLY SLoc WISE
# ==========================================================
# Covers:
# - Individual SLoc
# - Monthly from 2020 to 2026
# ==========================================================

monthly_sloc = (
    df.groupby(['SLoc', 'Year', 'Month'])
    .agg(
        Avg_Temp=('Tavg', 'mean'),
        Failure_Count=('Failure_Count', 'sum')
    )
    .reset_index()
)

print("\nMONTHLY SLoc WISE")
print(monthly_sloc)

# ==========================================
# OPTIONAL:
# SORT DATAFRAMES
# ==========================================

monthly_india = monthly_india.sort_values(
    by=['Year', 'Month']
)

monthly_region = monthly_region.sort_values(
    by=['Region', 'Year', 'Month']
)

monthly_sloc = monthly_sloc.sort_values(
    by=['SLoc', 'Year', 'Month']
)

# ==========================================
# OPTIONAL:
# EXPORT TO EXCEL
# ==========================================

with pd.ExcelWriter("Monthly_Temp_vs_Failure_Datasets.xlsx") as writer:

    monthly_india.to_excel(
        writer,
        sheet_name='Monthly_India',
        index=False
    )

    monthly_region.to_excel(
        writer,
        sheet_name='Monthly_Region',
        index=False
    )

    monthly_sloc.to_excel(
        writer,
        sheet_name='Monthly_SLoc',
        index=False
    )

print("\nDatasets exported successfully!")
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

# Replace with your file name
df = pd.read_excel("Monthly_Temp_vs_Failure_Datasets.xlsx",
                   sheet_name='Monthly_India')

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# PEARSON CORRELATION
# ==========================================

corr, p_value = pearsonr(
    df['Avg_Temp'],
    df['Failure_Count']
)

# ==========================================
# PRINT RESULTS
# ==========================================

print("\n========== PEARSON CORRELATION ==========")

print(f"Pearson Correlation Coefficient : {corr:.4f}")
print(f"P-value                         : {p_value:.6f}")

# ==========================================
# INTERPRETATION
# ==========================================

if p_value < 0.05:
    print("\nResult: Statistically Significant Relationship")
else:
    print("\nResult: No Statistically Significant Relationship")

# Correlation strength interpretation
abs_corr = abs(corr)

if abs_corr < 0.2:
    strength = "Very Weak"
elif abs_corr < 0.4:
    strength = "Weak"
elif abs_corr < 0.6:
    strength = "Moderate"
elif abs_corr < 0.8:
    strength = "Strong"
else:
    strength = "Very Strong"

print(f"Correlation Strength            : {strength}")

# Direction interpretation
if corr > 0:
    print("Relationship Direction          : Positive")
elif corr < 0:
    print("Relationship Direction          : Negative")
else:
    print("Relationship Direction          : No Relationship")

# ==========================================
# SCATTER PLOT
# ==========================================

plt.figure(figsize=(8,6))

plt.scatter(
    df['Avg_Temp'],
    df['Failure_Count']
)

plt.xlabel("Average Temperature")
plt.ylabel("Failure Count")
plt.title("Pearson Correlation: Temperature vs Failures")

plt.grid(True)

plt.show()
```

```python
import pandas as pd
from scipy.stats import f_oneway
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_excel(
    "Monthly_Temp_vs_Failure_Datasets.xlsx",
    sheet_name='Monthly_India'
)

# ==========================================
# CREATE TEMPERATURE BINS
# ==========================================

# Create temperature intervals automatically
temp_bins = pd.cut(df['Avg_Temp'], bins=4)

# Store interval labels
df['Temp_Range'] = temp_bins.astype(str)

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# CREATE GROUPS FOR ANOVA
# ==========================================

groups = []

for temp_range in df['Temp_Range'].unique():

    group_data = df[
        df['Temp_Range'] == temp_range
    ]['Failure_Count']

    groups.append(group_data)

# ==========================================
# PERFORM ONE-WAY ANOVA
# ==========================================

f_stat, p_value = f_oneway(*groups)

# ==========================================
# PRINT RESULTS
# ==========================================

print("\n========== ONE-WAY ANOVA ==========")

print(f"F-Statistic : {f_stat:.4f}")
print(f"P-value     : {p_value:.6f}")

# ==========================================
# INTERPRETATION
# ==========================================

if p_value < 0.05:
    print("\nResult: Significant Difference Exists")
    print("Failure counts differ across temperature ranges.")
else:
    print("\nResult: No Significant Difference Exists")
    print("Failure counts do not differ significantly across temperature ranges.")

# ==========================================
# GROUP STATISTICS
# ==========================================

print("\n========== GROUP MEANS ==========")

group_means = (
    df.groupby('Temp_Range')['Failure_Count']
    .mean()
)

print(group_means)

# ==========================================
# BOXPLOT
# ==========================================

plt.figure(figsize=(12,6))

df.boxplot(
    column='Failure_Count',
    by='Temp_Range',
    rot=15
)

plt.title("ANOVA: Failure Count vs Temperature Ranges")
plt.suptitle("")

plt.xlabel("Temperature Range")
plt.ylabel("Failure Count")

plt.grid(True)

plt.show()
```

```python
import pandas as pd
from scipy.stats import kruskal
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_excel(
    "Monthly_Temp_vs_Failure_Datasets.xlsx",
    sheet_name='Monthly_India'
)

# ==========================================
# CREATE TEMPERATURE RANGES
# ==========================================

# Automatically divide temperatures into 4 ranges
temp_bins = pd.cut(df['Avg_Temp'], bins=4)

# Convert intervals to readable strings
df['Temp_Range'] = temp_bins.astype(str)

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# CREATE GROUPS FOR KRUSKAL-WALLIS TEST
# ==========================================

groups = []

for temp_range in df['Temp_Range'].unique():

    group_data = df[
        df['Temp_Range'] == temp_range
    ]['Failure_Count']

    groups.append(group_data)

# ==========================================
# PERFORM KRUSKAL-WALLIS TEST
# ==========================================

h_stat, p_value = kruskal(*groups)

# ==========================================
# PRINT RESULTS
# ==========================================

print("\n========== KRUSKAL-WALLIS TEST ==========")

print(f"H-Statistic : {h_stat:.4f}")
print(f"P-value     : {p_value:.6f}")

# ==========================================
# INTERPRETATION
# ==========================================

if p_value < 0.05:
    print("\nResult: Significant Difference Exists")
    print("Failure counts differ significantly across temperature ranges.")
else:
    print("\nResult: No Significant Difference Exists")
    print("Failure counts do not differ significantly across temperature ranges.")

# ==========================================
# GROUP STATISTICS
# ==========================================

print("\n========== GROUP MEDIANS ==========")

group_medians = (
    df.groupby('Temp_Range')['Failure_Count']
    .median()
)

print(group_medians)

# ==========================================
# BOXPLOT
# ==========================================

plt.figure(figsize=(12,6))

df.boxplot(
    column='Failure_Count',
    by='Temp_Range',
    rot=15
)

plt.title("Kruskal-Wallis Test: Failure Count vs Temperature Ranges")
plt.suptitle("")

plt.xlabel("Temperature Range")
plt.ylabel("Failure Count")

plt.grid(True)

plt.show()
```

```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_excel(
    "Monthly_Temp_vs_Failure_Datasets.xlsx",
    sheet_name='Monthly_India'
)

# ==========================================
# CREATE TEMPERATURE RANGES
# ==========================================

# Divide temperatures into 4 ranges
temp_bins = pd.cut(df['Avg_Temp'], bins=4)

# Convert intervals into readable labels
df['Temp_Range'] = temp_bins.astype(str)

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# TUKEY HSD TEST
# ==========================================

tukey = pairwise_tukeyhsd(
    endog=df['Failure_Count'],
    groups=df['Temp_Range'],
    alpha=0.05
)

# ==========================================
# PRINT RESULTS
# ==========================================

print("\n========== TUKEY HSD TEST ==========")

print(tukey)

# ==========================================
# CONVERT RESULTS TO DATAFRAME
# ==========================================

tukey_df = pd.DataFrame(
    data=tukey._results_table.data[1:],
    columns=tukey._results_table.data[0]
)

print("\n========== TUKEY RESULT DATAFRAME ==========")

print(tukey_df)

# ==========================================
# SIGNIFICANT COMPARISONS
# ==========================================

print("\n========== SIGNIFICANT COMPARISONS ==========")

significant = tukey_df[tukey_df['reject'] == True]

if len(significant) > 0:
    print(significant)
else:
    print("No statistically significant pairwise differences found.")

# ==========================================
# BOXPLOT
# ==========================================

plt.figure(figsize=(12,6))

df.boxplot(
    column='Failure_Count',
    by='Temp_Range',
    rot=15
)

plt.title("Tukey HSD: Failure Count vs Temperature Ranges")
plt.suptitle("")

plt.xlabel("Temperature Range")
plt.ylabel("Failure Count")

plt.grid(True)

plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

df = pd.read_excel(
    "Monthly_Temp_vs_Failure_Datasets.xlsx",
    sheet_name='Monthly_India'
)

# ==========================================
# CREATE TEMPERATURE RANGES
# ==========================================

# Divide Avg_Temp into 4 temperature ranges
temp_bins = pd.cut(df['Avg_Temp'], bins=4)

# Convert interval labels to string
df['Temp_Range'] = temp_bins.astype(str)

# ==========================================
# CREATE FAILURE CATEGORIES
# ==========================================
# Convert numerical failure counts into categories

failure_bins = pd.cut(
    df['Failure_Count'],
    bins=4,
    labels=[
        'Low Failure',
        'Moderate Failure',
        'High Failure',
        'Very High Failure'
    ]
)

df['Failure_Category'] = failure_bins

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# CREATE CONTINGENCY TABLE
# ==========================================

contingency_table = pd.crosstab(
    df['Temp_Range'],
    df['Failure_Category']
)

print("\n========== CONTINGENCY TABLE ==========")
print(contingency_table)

# ==========================================
# CHI-SQUARE TEST
# ==========================================

chi2, p_value, dof, expected = chi2_contingency(
    contingency_table
)

# ==========================================
# PRINT RESULTS
# ==========================================

print("\n========== CHI-SQUARE TEST ==========")

print(f"Chi-Square Statistic : {chi2:.4f}")
print(f"P-value              : {p_value:.6f}")
print(f"Degrees of Freedom   : {dof}")

# ==========================================
# INTERPRETATION
# ==========================================

if p_value < 0.05:
    print("\nResult: Significant Association Exists")
    print("Temperature range and failure category are associated.")
else:
    print("\nResult: No Significant Association Exists")
    print("Temperature range and failure category are independent.")

# ==========================================
# EXPECTED FREQUENCIES
# ==========================================

expected_df = pd.DataFrame(
    expected,
    index=contingency_table.index,
    columns=contingency_table.columns
)

print("\n========== EXPECTED FREQUENCIES ==========")
print(expected_df)

# ==========================================
# BOXPLOT
# ==========================================

plt.figure(figsize=(12,6))

df.boxplot(
    column='Failure_Count',
    by='Temp_Range',
    rot=15
)

plt.title("Chi-Square Analysis: Failure Count vs Temperature Ranges")
plt.suptitle("")

plt.xlabel("Temperature Range")
plt.ylabel("Failure Count")

plt.grid(True)

plt.show()
```

```python
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

# Load Region-wise monthly dataset
df = pd.read_excel(
    "Monthly_Temp_vs_Failure_Datasets.xlsx",
    sheet_name='Monthly_Region'
)

# ==========================================
# CLEAN REGION COLUMN
# ==========================================

df['Region'] = (
    df['Region']
    .str.strip()
    .str.replace(r'\s+', '', regex=True)
    .str.upper()
)
df['Temp_Range'] = pd.cut(
    df['Avg_Temp'],
    bins=4
)

# ==========================================
# CREATE TEMPERATURE RANGES
# ==========================================

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# STORE RESULTS
# ==========================================

all_results = []

# ==========================================
# REGIONWISE TUKEY HSD
# FAILURE COUNT vs TEMPERATURE RANGE
# ==========================================

regions = df['Region'].unique()

for region in regions:

    print(f"\n==============================")
    print(f"REGION : {region}")
    print(f"==============================")

    # ======================================
    # FILTER REGION DATA
    # ======================================

    region_df = df[df['Region'] == region]

    # ======================================
    # CHECK MINIMUM GROUPS
    # ======================================

    if region_df['Temp_Range'].nunique() < 2:
        print("Not enough temperature groups")
        continue

    # ======================================
    # TUKEY HSD TEST
    # ======================================

    tukey = pairwise_tukeyhsd(
        endog=region_df['Failure_Count'],
        groups=region_df['Temp_Range'],
        alpha=0.05
    )

    # ======================================
    # PRINT RESULTS
    # ======================================

    print("\nTukey HSD Results:")
    print(tukey)

    # ======================================
    # CONVERT TO DATAFRAME
    # ======================================

    tukey_df = pd.DataFrame(
        data=tukey._results_table.data[1:],
        columns=tukey._results_table.data[0]
    )

    # Add region column
    tukey_df['Region'] = region

    # Store results
    all_results.append(tukey_df)

    # ======================================
    # SIGNIFICANT COMPARISONS
    # ======================================

    significant = tukey_df[
        tukey_df['reject'] == True
    ]

    print("\nSignificant Comparisons:")

    if len(significant) > 0:
        print(significant)
    else:
        print("No significant pairwise differences found.")

    # ======================================
    # BOXPLOT
    # ======================================

    plt.figure(figsize=(10,6))

    region_df.boxplot(
        column='Failure_Count',
        by='Temp_Range',
        rot=15
    )

    plt.title(
        f"Tukey HSD: Failure Count vs Temperature Range\n{region}"
    )

    plt.suptitle("")

    plt.xlabel("Temperature Range")
    plt.ylabel("Failure Count")

    plt.grid(True)

    plt.show()

# ==========================================
# COMBINE ALL RESULTS
# ==========================================

if len(all_results) > 0:

    final_results = pd.concat(
        all_results,
        ignore_index=True
    )

    # ======================================
    # PRINT FINAL RESULTS
    # ======================================

    print("\n========== FINAL TUKEY HSD RESULTS ==========")

    print(final_results)

    # ======================================
    # EXPORT RESULTS
    # ======================================

    final_results.to_excel(
        "Regionwise_Tukey_HSD_Results.xlsx",
        index=False
    )

    print("\nResults exported successfully!")

else:
    print("\nNo valid Tukey HSD results generated.")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

# ==========================================
# LOAD DATASET
# ==========================================

# Load Region-wise monthly dataset
df = pd.read_excel(
    "Monthly_Temp_vs_Failure_Datasets.xlsx",
    sheet_name='Monthly_Region'
)

# ==========================================
# CLEAN REGION COLUMN
# ==========================================

df['Region'] = (
    df['Region']
    .astype(str)
    .str.strip()
    .str.replace(r'\s+', '', regex=True)
    .str.upper()
)

# ==========================================
# CREATE TEMPERATURE RANGES
# ==========================================

df['Temp_Range'] = pd.cut(
    df['Avg_Temp'],
    bins=4,
    precision=1
)

# ==========================================
# CREATE FAILURE CATEGORIES
# ==========================================

df['Failure_Category'] = pd.cut(
    df['Failure_Count'],
    bins=4,
    labels=[
        'LOW',
        'MODERATE',
        'HIGH',
        'VERY_HIGH'
    ]
)

# ==========================================
# DISPLAY DATA
# ==========================================

print(df.head())

# ==========================================
# STORE RESULTS
# ==========================================

results = []

# ==========================================
# REGIONWISE CRAMER'S V
# FAILURE CATEGORY vs TEMPERATURE RANGE
# ==========================================

regions = df['Region'].unique()

for region in regions:

    print(f"\n==============================")
    print(f"REGION : {region}")
    print(f"==============================")

    # ======================================
    # FILTER REGION DATA
    # ======================================

    region_df = df[df['Region'] == region]

    # ======================================
    # CREATE CONTINGENCY TABLE
    # ======================================

    contingency_table = pd.crosstab(
        region_df['Temp_Range'],
        region_df['Failure_Category']
    )

    print("\nContingency Table:")
    print(contingency_table)

    # ======================================
    # CHECK TABLE SIZE
    # ======================================

    if contingency_table.shape[0] < 2 or contingency_table.shape[1] < 2:
        print("Not enough categories for Cramer's V")
        continue

    # ======================================
    # CHI-SQUARE TEST
    # ======================================

    chi2, p_value, dof, expected = chi2_contingency(
        contingency_table
    )

    # ======================================
    # CALCULATE CRAMER'S V
    # ======================================

    n = contingency_table.sum().sum()

    min_dimension = min(
        contingency_table.shape[0] - 1,
        contingency_table.shape[1] - 1
    )

    cramers_v = np.sqrt(
        chi2 / (n * min_dimension)
    )

    # ======================================
    # INTERPRETATION
    # ======================================

    if p_value < 0.05:
        significance = "Significant"
    else:
        significance = "Not Significant"

    # ======================================
    # CRAMER'S V STRENGTH
    # ======================================

    if cramers_v < 0.1:
        strength = "Negligible"
    elif cramers_v < 0.3:
        strength = "Weak"
    elif cramers_v < 0.5:
        strength = "Moderate"
    else:
        strength = "Strong"

    # ======================================
    # STORE RESULTS
    # ======================================

    results.append({
        'Region': region,
        'Chi2_Statistic': round(chi2, 4),
        'P_Value': round(p_value, 6),
        'Degrees_of_Freedom': dof,
        'Cramers_V': round(cramers_v, 4),
        'Association_Strength': strength,
        'Significance': significance
    })

    # ======================================
    # PRINT RESULTS
    # ======================================

    print(f"\nChi-Square Statistic : {chi2:.4f}")
    print(f"P-value              : {p_value:.6f}")
    print(f"Degrees of Freedom   : {dof}")
    print(f"Cramer's V           : {cramers_v:.4f}")
    print(f"Association Strength : {strength}")
    print(f"Result               : {significance}")

    # ======================================
    # EXPECTED FREQUENCIES
    # ======================================

    expected_df = pd.DataFrame(
        expected,
        index=contingency_table.index,
        columns=contingency_table.columns
    )

    print("\nExpected Frequencies:")
    print(expected_df)

    # ======================================
    # BOXPLOT
    # ======================================

    plt.figure(figsize=(10,6))

    region_df.boxplot(
        column='Failure_Count',
        by='Temp_Range',
        rot=15
    )

    plt.title(
        f"Cramer's V: Failure Count vs Temperature Range\n{region}"
    )

    plt.suptitle("")

    plt.xlabel("Temperature Range")
    plt.ylabel("Failure Count")

    plt.grid(True)

    plt.show()

# ==========================================
# FINAL RESULTS DATAFRAME
# ==========================================

results_df = pd.DataFrame(results)

# ==========================================
# PRINT FINAL RESULTS
# ==========================================

print("\n========== REGIONWISE CRAMER'S V RESULTS ==========")

print(results_df)

# ==========================================
# EXPORT RESULTS
# ==========================================

results_df.to_excel(
    "Regionwise_Cramers_V_Results.xlsx",
    index=False
)

print("\nResults exported successfully!")
```
